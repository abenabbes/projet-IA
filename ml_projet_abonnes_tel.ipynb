```python
# Un ensemble d'import
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
# Normalisation des données
from sklearn.preprocessing import MinMaxScaler
# diviser des données en entrainement, Test e Validation
from sklearn.model_selection import train_test_split
# Rechantionner les données
from sklearn.utils import resample
# Ensemble de model à tester
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
# Recherche de l'hyperparamétre de lalgorithme
from sklearn.model_selection import GridSearchCV
# Métriques pour l'évaluation
from sklearn.metrics import classification_report
# Pour éduire la dimention des données
from sklearn.feature_selection import RFE
```

```python
URL = 'https://raw.githubusercontent.com/JosueAfouda/TUTORIALS/main/Telco-Customer-Churn.csv'
df = pd.read_csv(URL)
df.head()
```

```python
categorie_culums = df.nunique()[df.nunique() < 5 ].keys().to_list()
categorie_culums
```

```python
# Création d'une fonction pour afficher un graphique à bar
# pour visualiser et analyser les données un graphique ieterractif
def affichage_graphe(a):
  df[a].value_counts(normalize=True).plot(kind='bar')
  plt.ylabel('Proportion')
  plt.title('Distribution de : ' + str(a))
  return plt.show()

interact(affichage_graphe, a = categorie_culums);
```
